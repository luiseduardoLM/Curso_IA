```python
# Importo librerias 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

# Cargo el archiuvo csv

df = pd.read_csv("wine_data.csv")

df.head(10)
```

```python
class_count = df["class"].value_counts()
print(class_count)

class_count.plot(kind="bar", color=["red", "blue", "pink"])
plt.title("Conteo de clases")
plt.xlabel("Clase")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
alcohol_count = df["Alcohol"].value_counts()
print(alcohol_count)

alcohol_count.plot(kind="bar")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
proline_count = df["Proline"].value_counts()
print(proline_count)

proline_count.plot(kind="bar")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
