```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv")

df.head(10)
```

```python
mean_sepal_lenght = df["sepal.length"].mean()
std_sepal_lenght = df["sepal.length"].std()
median_sepal_lenght = df["sepal.length"].median()

print("The mean is:", mean_sepal_lenght)
print("The standar desviation is:", std_sepal_lenght)
print("The median is:", median_sepal_lenght)
```
